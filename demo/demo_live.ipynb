# xOceania, live pond demo

**The story in one breath:** a pond looks fine all day. Overnight, oxygen quietly
falls and crashes below the lethal line just before dawn. The farmer only sees it
at sunrise, after the loss. xOceania's twin saw it coming hours earlier, and when
xOceania is allowed to act, the crash never happens.

This notebook is the **live fallback** for the pre-rendered video. Run it top to
bottom (Kernel, Restart and Run All); it finishes in well under 30 seconds and
draws the whole story on real output from the validated `xoceania_sim` model.

```python
import os, sys, time
# Make the demo modules importable whether the kernel started in the repo root
# or in demo/.
for cand in ('demo', '.', os.path.dirname(os.path.abspath('demo_live.ipynb'))):
    if os.path.isdir(cand) and cand not in sys.path:
        sys.path.insert(0, cand)
import xoceania_demo as demo
import render
from IPython.display import display, Video, Image
print('Imports OK. xoceania_sim model loaded.')
```

## 1. Simulate both ponds on identical weather

Same pond, same Mekong Delta forcing. The only difference is the control policy:
a typical daytime **aeration timer** (`do nothing`) versus xOceania's
**forecast-driven** aeration (`xOceania acts`).

```python
t0 = time.time()
data = demo.compute_demo()
b, x = data.do_nothing, data.xoceania
print(f'simulated both ponds in {time.time() - t0:.1f} s\n')
print(f'Do nothing : DO troughs at {data.trough_do:5.2f} mg/L at {data.trough_time:.0f} h (pre-dawn)')
print(f'xOceania   : DO never drops below {x.DO.min():5.2f} mg/L')
print(f'Forecast   : twin saw the crash {data.forecast_lead_h:.0f} h early '
      f'(predicted min DO {data.forecast_value:.2f} mg/L at {data.forecast_time:.0f} h)')
```

## 2. The whole story on one chart

The orange pond crashes through the 5 mg/L stress line into the shaded danger
zone and bottoms out pre-dawn (the farmer's loss). The vertical marker is where
the twin forecast the crash, 6 hours ahead. The teal pond, where xOceania acts,
holds above the line all night.

```python
fig, ax = render.build_contrast_figure(data)
fig.set_size_inches(13, 7.3)  # comfortable inline size
display(fig)
```
